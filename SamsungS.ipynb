```python
with open('SamsungS_trainning.csv', 'r') as f:
    df_new = pd.read_csv(f, sep=',', low_memory=False)
```

```python
with open('SamsungS_test.csv', 'r') as f:
    df_test = pd.read_csv(f, sep=',', low_memory=False)
```

```python
Xnew, ynew = df_new.iloc[:, : -1  ],df_new.iloc[:,-1]
```

```python
#Import Library
from sklearn import svm
# Create SVM classification object 
model = svm.SVC(kernel='linear', C = 1.0) 
```

```python
model.fit(Xnew,ynew)
```

```python
print(model.score(Xnew,ynew))
```

```python
model.predict(df_test)
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model,Xnew,ynew , cv=9 )
print("scores: %s mean: %f std %f" %(str(scores),np.mean(scores),np.std(scores)))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(random_state=1)
```

```python
rf.fit(Xnew,ynew)
```

```python
print(rf.score(Xnew,ynew))
```

```python
rf.predict(df_test)
```

```python
rfscores = cross_val_score(rf, Xnew, ynew, cv = 9)
print("random forest scores: %s  mean: %r  std: %f" % (str(rfscores), np.mean(rfscores), np.std(rfscores)))
```
